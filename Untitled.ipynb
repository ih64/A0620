##Align Images
before this step, I vetted the images by eye and threw out any that were pathologically bad
the images are under the sub directories fitsimages/B, fitsimages/V, fitsimages/I
for each filter you have to choose a reference image. all the others will be aligned with respect to the reference
by hand, i chose images on days were the seeing was pretty good to be the reference images. 

````python
#this takes a long time to run bc there are many images. just be patient and let the program do its thing
#i would also recommend running this on one of the department's heavier duty computers, i.e. tortoise, turtle, etc
#the aligned images are output to a new directory for each filter. ie aligned Iband images go to fitsimages/I_align

#this dictionary contains the reference image for each filter
refdict={"B":"rccd120315.0012.fits","V":"rccd120315.0013.fits","I":"rccd120315.0014.fits"}

#loop over the different filters
for i in ["B","V","I"]:
    #get a listing of the path for each image for a given filter
    images_to_align = sorted(glob.glob("fitsimages/"+i+"/*.fits"))
    ref_image = "fitsimages/"+i+"/"+refdict[i]

    #everything below here i copied from the alipy demo http://obswww.unige.ch/~tewes/alipy/tutorial.html
    identifications = alipy.ident.run(ref_image, images_to_align, visu=False)
    # That's it !
    # Put visu=True to get visualizations in form of png files (nice but much slower)
    # On multi-extension data, you will want to specify the hdu (see API doc).

    # The output is a list of Identification objects, which contain the transforms :
    for id in identifications: # list of the same length as images_to_align.
            if id.ok == True: # i.e., if it worked

                    print "%20s : %20s, flux ratio %.2f" % (id.ukn.name, id.trans, id.medfluxratio)
                    # id.trans is a alipy.star.SimpleTransform object. Instead of printing it out as a string,
                    # you can directly access its parameters :
                    #print id.trans.v # the raw data, [r*cos(theta)  r*sin(theta)  r*shift_x  r*shift_y]
                    #print id.trans.matrixform()
                    #print id.trans.inverse() # this returns a new SimpleTransform object

            else:
                    print "%20s : no transformation found !" % (id.ukn.name)

    # Minimal example of how to align images :

    outputshape = alipy.align.shape(ref_image)
    # This is simply a tuple (width, height)... you could specify any other shape.

    for id in identifications:
            if id.ok == True:
                    # Variant 2, using geomap/gregister, correcting also for distortions :
                    alipy.align.irafalign(id.ukn.filepath, id.uknmatchstars, id.refmatchstars,verbose=False,
                                          shape=outputshape, makepng=False, outdir="fitsimages/"+i+"_align/")
    print "finished "+i
````

##Read Photometry Data into Pandas Data Frames
in the previous section, we created many output files. now we need to read in the data from the output photometry files into a pandas data frame. that way we can efficently and easily work on the data. 

````python
#function reads open photometry files and looks up the quantities we're interested in, like magnitude, error, jd, airmass
#then takes this data and puts it in a pandas data frame
def magtodf(pathToMagFiles):
    magfiles=glob.glob(pathToMagFiles)
    
    #the keys for this dictionary will be the columns for the data frame.
    #each key starts off with an empy python list as its value.
    #we will read each output file one by one and append its data to the appropriate lists
    #so that each element in the list for a key will represents a row in that column 
    rowdict={'fname':[],'date':[],'juliandate':[],'airmass':[],
             'mag1':[],'mag2':[],'mag3':[],'mag4':[],'mag5':[],
             'merr1':[], 'merr2':[], 'merr3':[], 'merr4':[], 'merr5':[]}

    for i in magfiles:
        #use the handy astropy ascii object to read photometry files
        photdata=ascii.read(i)
        #the astropy.ascii automatically masks any bad data values. I'd prefer to keep the bad values as np.nans
        if np.ma.is_masked(photdata['MAG']):
            photdata['MAG'][photdata['MAG'].mask]=np.nan
        if np.ma.is_masked(photdata['MERR']):
            photdata['MERR'][photdata['MERR'].mask]=np.nan
        #for append the data from this photometry file to the dictionary
        rowdict['fname'].append(photdata['IMAGE'][0])
        rowdict['date'].append(float(photdata['IMAGE'][0][4:10]))
        rowdict['juliandate'].append(photdata['OTIME'][0])
        rowdict['airmass'].append(photdata['XAIRMASS'][0])
        rowdict['mag1'].append(photdata['MAG'][0])
        rowdict['mag2'].append(photdata['MAG'][1])
        rowdict['mag3'].append(photdata['MAG'][2])
        rowdict['mag4'].append(photdata['MAG'][3])
        rowdict['mag5'].append(photdata['MAG'][4])
        rowdict['merr1'].append(photdata['MERR'][0])
        rowdict['merr2'].append(photdata['MERR'][1])
        rowdict['merr3'].append(photdata['MERR'][2])
        rowdict['merr4'].append(photdata['MERR'][3])
        rowdict['merr5'].append(photdata['MERR'][4])
    
    #create a pandas dataframe out of the dictionary
    df=pd.DataFrame(rowdict)
    #change the data type of the date column to int. cant translate from string to int above for some reason
    df['date']=df['date'].astype(int)
    return df
````

##Calibration and Clean Up
calibrate the photometry of A0620 by calculating the offset for our comparison stars. Michelle looked up their magnitudes in the UCAC4 catalog. We can find the average difference between the instrumental magnitudes for the comparison stars and thier magnitudes in UCAC4 and use this as the offset. we then apply this offset to the instrumental magnitude of A0620 to transform its instrumental magnitude to the true magnitude scale

we will aslo find any outling points and remove them. When we are satisfied with the data set, we will dump the calibrated photometry to csv files for future reference.

````python
#subtract the true magnitudes that michelle looked up in UCAC4 for these comparison stars to get an offset
#then apply the offset to the xrb to get a calibrated magnitude
photBTableMC['offset']=(photBTableMC.ix[:,['mag2','mag3','mag4', 'mag5']] - np.array([15.741 ,16.185, 16.142, 17.750])).mean(axis=1)
photBTableMC['a0620Cal']=photBTableMC['mag1']-photBTableMC['offset']

photVTableMC['offset']=(photVTableMC.ix[:,['mag2','mag3','mag4', 'mag5']] - np.array([14.658, 15.353, 15.375, 16.378 ])).mean(axis=1)
photVTableMC['a0620Cal']=photVTableMC['mag1']-photVTableMC['offset']

photITableMC['offset']=(photITableMC.ix[:,['mag2','mag3','mag4', 'mag5']] - np.array([13.91, 14.23, 14.14, 14.03 ])).mean(axis=1)
photITableMC['a0620Cal']=photITableMC['mag1']-photITableMC['offset']
````

````python
#function takes a data frame made above, filter, period (in days) and option to color the points
#and plots a period folded light curve and more normal light curve 
def phaseAndlightCurve(dataframe,flt,t,color='r'):
    fig=plt.figure(figsize=(15,8))
    ax1=fig.add_subplot(211)

    ax1.plot((dataframe['juliandate'].values.astype(float)%t)/(t),
            dataframe['a0620Cal'],color+'o',ms=5)

    ax1.set_xlabel('Phase')
    ax1.set_ylabel(flt+' mag')

    for i in ([ax1.xaxis.label, ax1.yaxis.label] + ax1.get_xticklabels() + ax1.get_yticklabels()):
        i.set_fontsize(15)

    ax1.invert_yaxis()

    ax2=fig.add_subplot(212)

    ax2.errorbar(dataframe['juliandate'].astype(float)-2450000.5, dataframe['a0620Cal'], yerr=dataframe.merr1, fmt=color+'o')
    ax2.set_xlabel('MJD')
    ax2.set_ylabel(flt+' mag')

    for i in ([ax2.xaxis.label, ax2.yaxis.label] + ax2.get_xticklabels() + ax2.get_yticklabels()):
        i.set_fontsize(15)
    ax2.invert_yaxis()
    plt.show()
````

##Combined Data Set from Andy Cantrel and Rachel McDonald

Rachel gave me the data set Andy and she put together for A0620. They used  different comparison stars than Michelle, so there may be some systematic differences in the light curves. lets see if this is the case, and if so, how we can remove any systematic errors.

````python
def residualsRMAC(flt,t):
    IH=pd.read_csv('A0620_'+flt+'.csv')
    RMAC=pd.read_table('a0620oldersmartsdata/'+flt+'SMARTS', sep='\s+', names=['juliandate','mag'])
    #we have to force the formats and datatypes to match if we are going to get the join to match
    RMAC=RMAC.apply(lambda x: x.astype(float), axis=1).applymap(lambda x: '%.4f' % x)
    IH=IH.apply(lambda x: x.astype(float), axis=1).applymap(lambda x: '%.4f' % x)
    
    #do a dataframe join on the julian date with the old data and new data
    #this will produce a new data frame, 'overlap', which has columns for old mag and new mags
    #at the intersection of the data sets. then we can compare data night by night
    overlap=pd.merge(IH, RMAC, on='juliandate', how='inner')
    fig=plt.figure(figsize=(15,8))
    ax1=fig.add_subplot(311)

    ax1.plot(overlap.juliandate, overlap.a0620Cal,'ro')
    ax1.plot(overlap['juliandate'], overlap['mag'],'bo')

    ax2=fig.add_subplot(312)
    ax2.plot(overlap.juliandate, (overlap.a0620Cal.astype(float) - overlap['mag'].astype(float)), 'go')
    
    #also overplot the folded light curve from both data sets
    #if there are systematic differences here, it will hinder our ability to tell
    #if the A0620 is in the active or passive state
    ax3=fig.add_subplot(313)
    ax3.plot((overlap.juliandate.values.astype(float)%t)/(t), overlap.mag, 'bo')
    ax3.plot((overlap.juliandate.values.astype(float)%t)/(t), overlap.a0620Cal, 'ro')
    
    print "RMS: ", np.sqrt((np.square(overlap.a0620Cal.astype(float) - overlap['mag'].astype(float))).mean())
    print "mean abs val of residual: ", abs((overlap.a0620Cal.astype(float) - overlap['mag'].astype(float))).mean()
    print "mean difference: ", (overlap.a0620Cal.astype(float) - overlap['mag'].astype(float)).mean()
````

````python
#function reads open photometry files and looks up the quantities we're interested in, like magnitude, error, jd, airmass
#then takes this data and puts it in a pandas data frame
#This is a redefenition of the function, so that it is hard coded to look for data from 3 stars
#Ideally, it would figure out how many stars are in the mag files, but I was lazy and just re-wrote it
def magtodf(pathToMagFiles):
    magfiles=glob.glob(pathToMagFiles)
    
    rowdict={'fname':[],'date':[],'juliandate':[],'airmass':[],
             'mag1':[],'mag2':[],'mag3':[],
             'merr1':[], 'merr2':[], 'merr3':[]}

    for i in magfiles:
        photdata=ascii.read(i)
        if np.ma.is_masked(photdata['MAG']):
            photdata['MAG'][photdata['MAG'].mask]=np.nan
        if np.ma.is_masked(photdata['MERR']):
            photdata['MERR'][photdata['MERR'].mask]=np.nan
        rowdict['fname'].append(photdata['IMAGE'][0])
        rowdict['date'].append(float(photdata['IMAGE'][0][4:10]))
        rowdict['juliandate'].append(photdata['OTIME'][0])
        rowdict['airmass'].append(photdata['XAIRMASS'][0])
        rowdict['mag1'].append(photdata['MAG'][0])
        rowdict['mag2'].append(photdata['MAG'][1])
        rowdict['mag3'].append(photdata['MAG'][2])

        rowdict['merr1'].append(photdata['MERR'][0])
        rowdict['merr2'].append(photdata['MERR'][1])
        rowdict['merr3'].append(photdata['MERR'][2])

    
    df=pd.DataFrame(rowdict)
    #change the data type of the date column to int. cant translate from string to int above for some reason
    df['date']=df['date'].astype(int)
    return df
````

````python
#make photometry tables for B, V, I
photBTableAC=magtodf('fitsimages/B_align/*mag.AC')
photVTableAC=magtodf('fitsimages/V_align/*mag.AC')
photITableAC=magtodf('fitsimages/I_align/*mag.AC')
````

````python
#hacky routine I wrote to make a three panel plot of the new data, and old data
#but only at the intersection of the data sets-where they contain observations from identical nights
#top pannel: light curves from both data sets, middle pannel: residuals of data sets,
#bottom pannel: folded light curves of both data sets. Helps look for systematic offsets
#also prints out the mean of residuals, typical residual value, and RMS
def residualsRMAC2(flt,t):
    IH=pd.read_csv('A0620_AC_'+flt+'.csv')
    RMAC=pd.read_table('a0620oldersmartsdata/'+flt+'SMARTS', sep='\s+', names=['juliandate','mag'])
    #we have to force the formats and datatypes to match if we are going to get the join to match
    RMAC=RMAC.apply(lambda x: x.astype(float), axis=1).applymap(lambda x: '%.4f' % x)
    IH=IH.apply(lambda x: x.astype(float), axis=1).applymap(lambda x: '%.4f' % x)
    
    #do a dataframe join on the julian date with the old data and new data
    #this will produce a new data frame, 'overlap', which has columns for old mag and new mags
    #at the intersection of the data sets. then we can compare data night by night
    overlap=pd.merge(IH, RMAC, on='juliandate', how='inner')
    fig=plt.figure(figsize=(15,8))
    ax1=fig.add_subplot(311)

    ax1.plot(overlap.juliandate, overlap.A0620Cal,'ro')
    ax1.plot(overlap['juliandate'], overlap['mag'],'bo')

    ax2=fig.add_subplot(312)
    ax2.plot(overlap.juliandate, (overlap.A0620Cal.astype(float) - overlap['mag'].astype(float)), 'go')
    
    #also overplot the folded light curve from both data sets
    #if there are systematic differences here, it will hinder our ability to tell
    #if the A0620 is in the active or passive state
    ax3=fig.add_subplot(313)
    ax3.plot((overlap.juliandate.values.astype(float)%t)/(t), overlap.mag, 'bo')
    ax3.plot((overlap.juliandate.values.astype(float)%t)/(t), overlap.A0620Cal, 'ro')
    
    print "RMS: ", np.sqrt((np.square(overlap.A0620Cal.astype(float) - overlap['mag'].astype(float))).mean())
    print "mean abs val of residual: ", abs((overlap.A0620Cal.astype(float) - overlap['mag'].astype(float))).mean()
    print "mean difference: ", (overlap.A0620Cal.astype(float) - overlap['mag'].astype(float)).mean()
````

# Nights With Multiple Observations
ATel #7547 has a plot that shows nights in 2009, 2011, and 2013 where A0620 was intenesly monitored. There is some flickering and short term variability in the target over the course ove these observations. I'm interested to see if our data set contains nights with many observations

````python
def getJDS(flt):
    #grab the new data and old data, and concat them ontop of each other into one pandas data frame
    IHdata=pd.read_csv('A0620_AC_'+flt+'.csv', index_col=False).drop(['Unnamed: 0','merr1'], axis=1)
    IHdata['id']=np.ones(len(IHdata))
    ACdata=pd.read_table('a0620oldersmartsdata/'+flt+'SMARTS', sep='\s+', names=['juliandate','A0620Cal'])
    ACdata['id']=np.zeros(len(ACdata))
    alldata=pd.concat([ACdata,IHdata])
    
    #grab the julain dates, force them to floats, chop off the decimal places,
    #flatten out the array, save numpy array in variable JDS
    JDS=alldata.ix[:,['juliandate']].applymap(lambda x: '%.4f' % x).values.astype(float).ravel()
    
    #sort the array
    return np.sort(JDS)
    
def findRepeat(dateArray,threshold):    
    repeatDates={}
    i=0
    j=i+1
    n=len(dateArray)
    
    #use pointers to progress down the array
    #if you find repeats, add them to the dictionary along with their frequency
    while j < n:
        datenow=dateArray[i]
        nextdate=dateArray[j]
        if nextdate-datenow < .5:
            if datenow in repeatDates:
                repeatDates[datenow]+=1
            else:
                repeatDates[datenow]=2
            j+=1
        else:
            i=j
            j+=1
    
    trimrepeatDates={}
    
    for key in repeatDates:
        if repeatDates[key] >= threshold:
            trimrepeatDates[key]=repeatDates[key]
    
    return trimrepeatDates
````
